```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Dados/IceCreamData.csv')
```

```python
df['Temperature'] = df['Temperature'].astype(int)
df['Revenue'] = df['Revenue'].astype(int)
result = df.sort_values('Temperature')
result
plt.title('Ice Cream Truck')
plt.plot(result['Temperature'], result['Revenue'],'r.--', label = 'US$')
plt.xlabel('Temperature(Fahrenheit)')
plt.ylabel('Sales(US$)')
plt.legend()
plt.show()
#plt.savefig('ResultadoFinal.png', dpi=300, bbox_inches='tight')
```
