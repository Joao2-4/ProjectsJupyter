```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv('Dados/gas_prices.csv')
df
```

```python
plt.title('Crescimento do valor do combustivel em US$')

list_country = ['Australia','Italy','Mexico','France','Germany','USA']
colors = ['purple', 'blue', 'black', 'green', 'orange', 'red']
color_dict = dict(zip(list_country, colors))
for country in df:
    if country in list_country:
        plt.plot(df['Year'], df[country], label=country, marker='.', color=color_dict[country])   
       

#plt.plot(df['Year'], df['Australia'], '.-', label = 'Australia')
#plt.plot(df['Year'], df['Italy'], '.-', label = 'Italy')
#plt.plot(df['Year'], df['USA'], '.-', label = 'USA')
#plt.plot(df['Year'], df['Japan'],'.-', label = 'Japan')
#'r.-', <- vermelho e com pontinhos na linha, se fosse -- ficaria tracejado

plt.xticks(df['Year'][::2])

plt.xlabel('Years')
plt.ylabel('US$')
plt.legend()
plt.show()
#plt.savefig('GraficoCombustivel.png', dpi=300, bbox_inches='tight')
```
