```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Dados/titanic.csv')
```

```python
df1 = df[['Name','Age','Sex','Pclass','Survived']]
df2 = df1.loc[(df1['Sex'] == 'female') & (df1['Pclass'] == 1) & (df1['Survived'] == 1)]
df3 = df2.sort_values(['Name']) 
```

```python
df1 = df[['PassengerId','Sex','Pclass','Survived']]
df1.loc[df1['Pclass']== 1,['Pclass']] = '1a. Classe'
df1.loc[df1['Pclass']== 2,['Pclass']] = '2a. Classe'
df1.loc[df1['Pclass']== 3,['Pclass']] = '3a. Classe'
df1.loc[df1['Sex']== 'male',['Sex']] = 'H: '
df1.loc[df1['Sex']== 'female',['Sex']] = 'M: '
```

```python
plt.title('Total de pessoas no navio por Classe e gênero')
C1 = df1.groupby(['Sex', 'Pclass']).count().sort_values(['Pclass'], ascending=True) 
df_plot = C1.reset_index()
colors = ['lightsteelblue','pink']
plt.bar(df_plot['Sex']+ df_plot['Pclass'].astype(str), df_plot['PassengerId'], color=colors, label = 'Homem') 
plt.bar(0,0, label = 'Mulher', color='pink') 

plt.xticks(rotation=90)
plt.xlabel('Classes')
plt.ylabel('Total de pessoas')
plt.legend()
#plt.savefig('Total.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
plt.title('Total de pessoas no navio por Classe, gênero + \nsobrevivente e mortos')
df1.loc[df1['Survived']== 0,['Survived']] = 'Morto '
df1.loc[df1['Survived']== 1,['Survived']] = 'Vivo '
C1 = df1.groupby(['Pclass', 'Survived','Sex']).count()
df_plot = C1.reset_index()
colors = ['lightsteelblue','pink']
plt.bar(df_plot['Sex'] + df_plot['Survived'].astype(str) + df_plot['Pclass'].astype(str), df_plot['PassengerId'],color=colors, label = 'Homem')
plt.bar(0,0, label = 'mulher', color = 'pink')
plt.xticks(rotation=90)
plt.xlabel('Classes')
plt.ylabel('Total de pessoas')
plt.legend(loc=2)
#plt.savefig('Total2.png', dpi=300, bbox_inches='tight')
plt.show()
```
